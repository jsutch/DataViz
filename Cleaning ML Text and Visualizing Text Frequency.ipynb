## Manual Tokenization

- in-word punctuation is preserved (armour-like)
- so are chunks of statements ('"What's' and 'thought.')


**Splitting by whitespace**

```python
filename = 'metamorphosis_clean.txt'
file = open(filename, 'rt')
text = file.read()
file.close()
# split into words by white space
words = text.split()
print(words[:100])
```

**or - only grab words with regex**

```python
# split based on words only
import re
words = re.split(r'\W+', text)
print(words[:100])
```

**Removing punctuation**

- import the python constant "punctuation"
- make a map of all the ascii chars with .maketrans() 
- "What's" has become "Whats" and "armor-like" becomes "armorlike".
- "dream." has become "dream"

```python
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in words]
print(stripped[:100])
```

**Normalizing case**
- Easily done with .lower() to recreate the words list

```python

words = [word.lower() for word in words]
```

```python
print(words[:100])
```

## Tokenization and cleaning with NLTK

**Splitting into sentences**

```python
# split into sentences
from nltk import sent_tokenize
sentences = sent_tokenize(text)
print(sentences[0])
```

**Split into words**


```python
# split into words
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
print(tokens[:100])
```

**Strip punctuation and numbers, et al**
- using .isalpha()

```python
# remove all tokens that are not alphabetic
words = [word for word in tokens if word.isalpha()]
print(words[:100])
```

**Filtering out stopwords**

- pulling out words that don't contribute to the deeper meaning of the phrase. (e.g. “the“, “a“, and “is“)

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
print(stop_words)
```

**Pipelining the text cleaning**

Pipeline:
    - Load the raw text.
    - Split into tokens.
    - Convert to lowercase.
    - Remove punctuation from each token.
    - Filter out remaining tokens that are not alphabetic.
    - Filter out tokens that are stop words.


```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
# convert to lower case
tokens = [w.lower() for w in tokens]
# remove punctuation from each word
import string
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in tokens]
# remove remaining tokens that are not alphabetic
words = [word for word in stripped if word.isalpha()]
# filter out stop words
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
words = [w for w in words if not w in stop_words]
print(words[:100])
```

**Grabbing stemmed words using PorterStemmer**

- break down words to their stems
- reducing each word to its root or base. (e.g. “fishing,” “fished,” “fisher” all reduce to the stem “fish.”)


```python
# stemming of words
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
stemmed = [porter.stem(word) for word in words]
print(stemmed[:100])
```

## Analyzing and visualizing the text

### Counting individual tokens

**making  dict of word counts**

```python
# using pre-stemmed tokens
wordcount = {}
for w in words:
    if w in wordcount:
        wordcount[w] += 1
    else:
        wordcount[w] = 1
```

pre-filtering the results
- sorting by count value
- turning the counts into a list, then slicing
- recombining output into a dict

```python
top50 = dict(list(sorted(wordcount.items(), key=lambda x: x[1], reverse=True))[:50])
```

```python
list(top50.items())[:10]
```

**barplotting the dict with pyplot.bar()**

```python
plt.figure(figsize=(15,6))
plt.xticks(rotation=45, ha='right')
plt.bar(top50.keys(), top50.values())
```

**Plotting Word Frequency with sns.countplot()**

Another method to pre-filter total results, but leave the number of tokens in place, is to set a floor for total terms. In this case we'll only be interested in words that appear over 20 times.

- use both the raw tokens list and the dictionary to create a new list to feed into .countplot()

```python
frequentwords = [w for w in words if wordcount[w] > 20]
```

```python
len(frequentwords)
```

```python
plt.figure(figsize=(25,10))
chart = sns.countplot(frequentwords)
chart.set_xticklabels(
    chart.get_xticklabels(), 
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'

)
plt.tight_layout(pad=0)
```

**Showing a wordcloud of the top 50 terms**

- wordcloud has a built-in limiter "max_words"

```python
plt.figure(figsize=(10,10))
wordcloud = WordCloud(background_color="white",width=1000,height=1000, max_words=50,relative_scaling=0.5,normalize_plurals=False).generate_from_frequencies(wordcount)
plt.imshow(wordcloud)
plt.axis("off"); # this needs a closing semicolon or it gets ignored
```

**Word cloud maps onto shapes**

Possibly not overly useful, but another option for display

```python
# modules for generating the word cloud
from os import path, getcwd
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
d = getcwd()
## join all documents in corpus
text = " ".join(words)
mask = np.array(Image.open(path.join(d, "globe.jpg")))
wc = WordCloud(background_color="white", max_words=1000, mask=mask,
               max_font_size=90, random_state=42)
wc.generate(text)
# create coloring from image
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[20,12])
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off");
_=plt.show()
```
